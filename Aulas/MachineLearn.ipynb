```python
# pacotes comuns
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
#Pacotes de machine learn
from sklearn.ensemble import RandomForestRegressor
from sklearn.datasets import fetch_openml
from sklearn.model_selection import train_test_split
boston = fetch_openml(name="boston", version=1)
print(boston);
```

```python
#Ignorar avisos
import warnings
warnings.filterwarnings("ignore");
```

```python
#X variavei explicativas
x = pd.DataFrame(boston.data, columns=boston.feature_names);
#Y Variavel de explicação
y = boston.target;
```

```python
print(x.shape);
x.head();
```

```python
print(y.shape);
```

```python
print(boston.DESCR);
```

```python
plt.figure(figsize = (12,8));
plt.scatter(x['RM'], y, c="b", alpha=0.6);
plt.title('Preço mediano e número de quartos', fontsize = 16);
plt.xlabel("Número médio de quartos no imovél", fontsize = 16);
plt.ylabel("Preço mediano  dos imovéis");
plt.show();
```

```python
plt.figure(figsize = (12,8));
plt.scatter(x['CRIM'], y, c="r", alpha=0.6);
plt.title('Preço mediano e número de criminalidade', fontsize = 16);
plt.xlabel("Crime per capta", fontsize = 16);
plt.ylabel("Preço mediano dos imovéis");
plt.show();
```

```python
model = RandomForestRegressor();
model.fit(x, y);
x["preco_previsto"] = model.predict(x);
```

```python
x.head();
print(x)
```

```python
plt.figure(figsize = (12,8));
plt.scatter(x['preco_previsto'], y, c="r", alpha=0.6);
plt.title('Preço mediano previsto e Preço mediano verdadeiro', fontsize = 16);
plt.xlabel("Preço mediano previsto", fontsize = 16);
plt.ylabel("Preço mediano verdadeiro");
plt.show();
```
